Spoken Languages

```python
import pandas as pd

# Load CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\ADMIN\Desktop\SJSU\SEM1\DB Analytics\Moviedataset\spoken_languages.csv")
df.head()
```

```python
import pandas as pd
import ast


# Function to extract values from JSON formatted column
def extract_values(json_str):
    try:
        data_list = ast.literal_eval(json_str)
        iso_639_1_list = [item['iso_639_1'] for item in data_list]
        name_list = [item['name'] for item in data_list]
        return ','.join(map(str, iso_639_1_list)), ','.join(map(str, name_list))
    except:
        return None, None

# Apply the function to each row of the DataFrame
df[['iso_639_1_values', 'name_values']] = df['spoken_languages'].apply(lambda x: pd.Series(extract_values(x)))

# Drop the original JSON formatted column
df.drop(columns=['spoken_languages'], inplace=True)

# Write the DataFrame to a new CSV file
df.to_csv(r'C:\Users\ADMIN\Desktop\SJSU\SEM1\DB Analytics\Moviedataset\spoken_languages1.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv(r'C:\Users\ADMIN\Desktop\SJSU\SEM1\DB Analytics\Moviedataset\spoken_languages1.csv')

# Function to split values in a cell and create new rows
def split_values(row):
    id_val = row['id']
    iso_639_1_values = str(row['iso_639_1_values']).split(',') if not pd.isnull(row['iso_639_1_values']) else []
    name_values = str(row['name_values']).split(',') if not pd.isnull(row['name_values']) else []
    rows = []
    for iso_639_1_values, name_value in zip(iso_639_1_values, name_values):
        rows.append([id_val, iso_639_1_values, name_value])
    return rows

# Apply the function to each row and flatten the resulting list of lists
new_rows = [row for sublist in df.apply(split_values, axis=1) for row in sublist]

# Create a new DataFrame from the flattened list
new_df = pd.DataFrame(new_rows, columns=['id', 'iso_639_1_values', 'name_values'])

# Export the DataFrame to a new CSV file
new_df.to_csv(r'C:\Users\ADMIN\Desktop\SJSU\SEM1\DB Analytics\Moviedataset\spoken_languages2.csv', index=False)
```

Belongs To Collection 

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv(r'C:\Users\ADMIN\Desktop\SJSU\SEM1\DB Analytics\Moviedataset\belongs_to_collection.csv')

# Function to extract values from JSON string
def extract_values(json_str):
    try:
        data = eval(json_str)  # Evaluate JSON string to dictionary
        return data.get('id', None), data.get('name', None), data.get('poster_path', None), data.get('backdrop_path', None)
    except:
        return None, None, None, None

# Apply the function to each row of the DataFrame
df[['collection_id', 'name', 'poster_path', 'backdrop_path']] = df['belongs_to_collection'].apply(lambda x: pd.Series(extract_values(x)))

# Drop the original JSON formatted column
df.drop(columns=['belongs_to_collection'], inplace=True)

# Write the DataFrame to a new CSV file
df.to_csv(r'C:\Users\ADMIN\Desktop\SJSU\SEM1\DB Analytics\Moviedataset\belongs_to_collection1.csv', index=False)
```
